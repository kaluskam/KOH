```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image
```

```python
def load_history(path):
    history = pd.read_csv(path, index_col=0)
    return history


def mean_scores(histories):
    n = len(histories)
    scores = np.zeros(len(histories[0]))
    for i in range(n):
        scores += histories[i]['adjusted rand score']
    return pd.DataFrame(scores / 5)
```
